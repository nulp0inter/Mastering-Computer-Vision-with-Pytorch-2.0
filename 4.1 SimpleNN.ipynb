```python
import torch
import torch.nn as nn
```

```python
class SimpleNN(nn.Module):
    
    def __init__(self):
        super(SimpleNN,self).__init__()
        
        #Defining Layers of NN and units per Layer
        #class torch.nn.Linear(in_features, out_features, bias=True, device=None, dtype=None)
        self.layer_1 = nn.Linear(3,5) # Input layer with 3 neurons and hidden layer with 5 neurons
        self.layer_2 = nn.Linear(5,1) # Hidden layer with 5 neurons and output layer with 1 neuron
             
        #Defining all Activation Function
        #class torch.nn.ReLU(inplace=False)
        self.activation = nn.ReLU() # Activation function
    
    def forward(self,x):
        x = self.layer_1(x)
        x = self.activation(x)
        x = self.layer_2(x)
        return x
```

```python
# Create an instance of the SimpleNN class
model = SimpleNN()  

# Generate random input data
input_data = torch.randn(1,3)

# Forward pass through the network
output = model(input_data)
print(output)
```
